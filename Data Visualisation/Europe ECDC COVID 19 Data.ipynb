```python
import numpy as np
import pandas as pd
# COVID 19 Europe cases from ECDC
data = pd.read_csv("https://opendata.ecdc.europa.eu/covid19/casedistribution/csv")
```

```python
# Selecting Ireland data from the list
dataIRL = data.loc[data['countriesAndTerritories'] == 'Ireland']
dataIRL
```

```python
import matplotlib.pyplot as plt

# Plot Cumulative_number_for_14_days_of_COVID-19_cases_per_100000
plt.figure(figsize=(10,10))
plt.plot(dataIRL["month"], dataIRL["Cumulative_number_for_14_days_of_COVID-19_cases_per_100000"])
plt.xlabel('Month',fontsize=20)
plt.ylabel('Cumulative_number_for_14_days_of_COVID-19_cases_per_100000',fontsize=20)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,10))
sns.lineplot(x=dataIRL.iloc[:,2], y=dataIRL.iloc[:,4], data=dataIRL, color="r")
plt.suptitle('Monthly Confrirmed cases in Ireland')
```

```python
plt.figure(figsize=(10,20))
g = sns.lineplot(x=data.iloc[:,2], y=data.iloc[:,4], data=data, hue ='countriesAndTerritories')
g.legend(loc='right', ncol=10)
```

```python
# COVID 19 Ireland Connaught-Ulster County Data
dataIRL = pd.read_csv("http://opendata-geohive.hub.arcgis.com/datasets/d9be85b30d7748b5b7c09450b8aede63_0.csv")
```

```python
dataIRL
```

```python
counties = ['Cavan','Galway','Letterkenny','Mayo','Sligo']
dataCU = dataIRL[dataIRL['CountyName'].isin(counties)] 
```

```python
# import datetime
from datetime import datetime
#Split the date and time
pd.to_datetime(dataCU["TimeStamp"]).dt.date
# Convert date to Month Day format. https://strftime.org/
pd.to_datetime(dataCU["TimeStamp"]).dt.strftime("%b %d") 

```

```python
import matplotlib.dates as mdates
fig = plt.figure(figsize=(15,10))

ax = plt.gca()

ax.xaxis.set_major_locator(mdates.DayLocator(interval=7))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d %b'))

groups = dataCU.groupby('CountyName')
for name, group in groups:    
    ax.scatter(group["TimeStamp"], group["ConfirmedCovidCases"], label = name, linestyle= "dotted")  
    
plt.setp(ax.get_xticklabels(), rotation=45, fontsize=15)
ax.grid(True)

plt.legend()
plt.show()
```
