# Histograms and Heatmaps

Histograms

```python
# The 'inline' flag will use the appropriate backend to make plots appear in the notebook
%matplotlib inline
import pandas as pd

# 'plt' is an alias for the 'matplotlib.pyplot' module
```

# Load data

```python
# The kc_house_data.csv from Kaggle is used for demonstration
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
# Make Histograms of the 'price' column
df['price'].hist()
```

```python
# Try changing the bin value to smaller intervals
df['price'].hist(bins=30)
```

```python
# Remove the outliers,which are a few houses with a price above 3 million and plot a histogram of the subset
price_subset = df.loc[:, 'price'] <= 3000000
df.loc[price_subset,'price'].hist(bins=30)
```

```python
# Add edgecolor to separate the bars
df.loc[price_subset,'price'].hist(bins=30, edgecolor = 'black')
```

# Heatmaps

```python
# Heatmap is a graphical representation of data where values are illustrated by colours
# We will use seaborn library
df = pd.read_csv('DropletData.csv')
df.head()
```

```python
# Sequential colourmap with Seaborn
# uninteresting values to relatively high interesting values
import matplotlib.pyplot as plt
import seaborn as sns

df = df.corr()

plt. figure(figsize = (6,6))
sns.heatmap(df, annot=True, cmap='Blues');
#plt.ylabel('maxFeret');
#plt.xlabel('minFeret');
```
